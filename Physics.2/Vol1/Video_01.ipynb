## Thermometers and Temperature Scales

### Zeroth Law of Thermodynamics

If bodies A and B are separately in thermal equilibrium with a third body C, then A and B are in thermal equilibrium with each other if placed in thermal contact.

Thermal equilibrium means that there isn't any heat transfer.


### Temperature Scales

Temperature measures the amount of Kinetic Energy that a substance has a molecular level.

Celsius Scale: 

* $0^{\circ}$ water at Ice Point
* $100^{\circ}$ water at Steam Point

Kelvin Scale:
* Same degree spacing as Celsius, but the absolute zero is the reference when the atomic motion stops (0 K), the reference sacle is $0^{\circ}$ Celsius = $273.15$ K.

* The conversion factor:
$$T_{kelvin} = T_{celsius} + 273.15 $$

Fahrenheit Scale:

* The conversion factor:
$$T_{fahrenheit} = \frac{9}{5}T_{celsius} + 32 $$





```python
temp_celsius = 0 * u.deg_C
temp_kelvin = temp_celsius.to(u.K, equivalencies=u.temperature())
temp_fahrenheit = temp_celsius.to(u.K, equivalencies=u.temperature())
temp_fahrenheit = temp_celsius.to(u.imperial.deg_F, equivalencies=u.temperature())

print(
    f"temperature: {temp_celsius} equivalent to {temp_kelvin} equitalent to {temp_fahrenheit}"
)
```

#### Problem 1

Convert $98.6^{\circ}$ F to celsius and kelvin?




```python
temp_f = 98.6 * u.imperial.deg_F
temp_c = temp_f.to(u.deg_C, equivalencies=u.temperature())
temp_k = temp_f.to(u.K, equivalencies=u.temperature())

print(f"temperature: {temp_f} equivalent to {temp_c:.0f} equitalent to {temp_k}")
```
