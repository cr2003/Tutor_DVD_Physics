## Thermal Expansion and And Contraction of Solids and Liquids

* One dimension:

<img src="../../pics/Phy2_Vol2_video_02.png" width=400> 


$$\Delta L = \alpha L_0 \Delta T$$

$\alpha$ is the coefficient of linear expansion, the units are $\left[ \frac{1}{^{\circ} C} \right]$


* Two dimensions:

<img src="../../pics/Phy2_Vol2_video_02_2.png" width=400> 

$$\Delta A = \gamma A_0 \Delta T$$

$\gamma$ is the coefficient of Area expansion


* Three dimensions:

$$\Delta V = \beta V_0 \Delta T$$

$\beta$ is the coefficient of Volume expansion


### Most Objects
* Contract when cooled
* Density increases as Temperature lowers
* The "frozen" liquid sinks because $\delta_{solid} > \delta_{liquid}$

### Water
* Contracts when cooled, but then expands near the freezing point
* The "ice" floats because $\delta_{solid} < \delta_{liquid}$

#### Problem 1

A steel bridge is 518 meters long. How much will its length change between temperature extres of -20 deg C and 35 deg C.




```python
L_0 = 518 * u.m
temp_low = -20 * u.deg_C
temp_high = 35 * u.deg_C
alpha_steal = 11e-6 * (1 / u.deg_C)

# calculate delta temp
delta_temp = temp_high - temp_low

# calculate delta L
delta_L = alpha_steal * L_0 * delta_temp

print(f"Delta L: {delta_L:.2f} equivalent to {delta_L.to(u.cm):.0f}")
```

#### Problem 2

A cylindrical brass sleeve is to fit over a brass pipe whose diameter is 3.212 cm at 0 deg C. The diameter of the sleeve is 3.196 cm at 0 deg C. Tho what temperature must the sleeve be heated before it will fit onto the pipe.


<img src="../../pics/Phy2_Vol2_video_02_3.png" width=400> 


```python
L_0 = 3.196 * u.cm
delta_L = (3.212 - 3.196) * u.cm
alpha_brass = 19e-6 * (1 / u.deg_C)

# calculate delta temp
# delta_L = alpha_brass * L_0 * delta_temp ==> delta_temp = delta_L / (alpha_brass * L_0)
delta_temp = delta_L / (alpha_brass * L_0)

print(f"Sleeve temp: {delta_temp:.1f}")
```

#### Problem 3

A tank is filled with gasoline to the brim with 45 liters of gas at 20 deg C. After this, the gas is heated to 35 deg C by the sun. How much gas overflows from the tank? Assume the tank itself is not affected by the sun.

<img src="../../pics/Phy2_Vol2_video_02_3.png" width=400> 


```python
V_0 = 45 * u.liter
T_0 = 10 * u.deg_C
T = 35 * u.deg_C
beta_gass = 9.6e-4 * (1 / u.deg_C)

# calculate delta temp
delta_temp = T - T_0
# calculate final Volume
delta_V = beta_gass * V_0 * delta_temp

print(f"Gas overflow: {delta_V:.2f}")
```
