##  Torque

#### Problem 1

The torque required to loosen a nut is 40 N*m. What force is required at the end of a 30 cm wrench to loosen the nut?


```python
tau = 40 * (u.N * u.m)
r = 30 * u.cm

# tau = r F ==> F = tau / r
F = tau / r.to(u.m)
print(f"Force: {F:.0f}")
```

#### Problem 2

A fishing pole is held at an angle of 20 deg and is 2 m in lenght. If a fish is on the end of the line and the line is 37 deg below the horizontal, how much torque dees the fish exert on the holder of the fishing pole? The fish is pooling with a force of 100 N.

<img src="../pics/Phy1_video_18_3.png" width=800> 



```python
angle_pole = 20 * u.deg
angle_line = 37 * u.deg
r = 2 * u.m
F = 100 * u.N

tau = r * F * np.sin(angle_pole + angle_line)
print(f"Torque: {tau:.0f}")
```
