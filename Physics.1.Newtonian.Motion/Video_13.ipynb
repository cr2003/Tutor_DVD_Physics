##  Elastic Collision

P & KE Conserved  (billiard balls)

Conservation of Momentum:

$$m_1 v_{1,i} + m_2 v_{2,i} =  m_1 v_{1,f} + m_2 v_{2,f}$$

Conservation of KE:

$$\frac{1}{2}m_1 v_{1,i}^2 + \frac{1}{2}m_2 v_{2,i}^2 =  \frac{1}{2}m_1 v_{1,f}^2 + \frac{1}{2}m_2 v_{2,f}^2$$


##  Inelastic Collision

Conservation of Momentum:

$$m_1 v_{1,i} + m_2 v_{2,i} =  m_1 v_{1,f} + m_2 v_{2,f}$$

##  Perfectly Inelastic Collision
The two object in the collision stick together resulting only one final compound object

Conservation of Momentum:

$$m_1 v_{1,i} + m_2 v_{2,i} =  (m_1 + m_2)  v_f$$

#### Problem 1

A skateboard (mass 1.2 kg) is rolling at a speed of 5 m/s and suddenly a 0.8 kg cat falls down from a tree onto the sakateboard. What is the speed of the skateboard after the cat lands?


```python
# -----------------------------------------------------------------
# 2. Numerical Evaluation with AstroPy Units
#
# Perfectly inelastic collision: m_1 v_{1,i} + m_2 v_{2,i} =  (m_1 + m_2) v_{f}
# -----------------------------------------------------------------

# Given parameters with AstroPy units
mass_skate = 1.2 * u.kg
mass_cat = 0.8 * u.kg

vel_skate = 5 * (u.m / u.s)
vel_cat = 0 * (u.m / u.s)


# conservation of momentum:  m_1 v_{1,i} + m_2 v_{2,i} =  (m_1 + m_2) v_{f}
# We only consider the x-direction
vel_final = ((mass_skate * vel_skate) + (mass_cat * vel_cat)) / (mass_skate + mass_cat)
print("--- AstroPy Numerical Result ---")
print(f"Final velocity: {vel_final:.0f}")
```

#### Problem 2 

A sphere (3 kg) makes a perfectly inelastic collision with a second sphere that is at rest. The final velocity of the 2 sphere system is equal to 1/3 the original speed of the 3 kg sphere. What is the mass of the 2nd sphere?


```python
# -----------------------------------------------------------------
# 1. Symbolic Solution with SymPy
# -----------------------------------------------------------------

# Define symbols for the algebraic solution
m_1, m_2, v_i, v_f = sp.symbols("m_1, m_2, v_i, v_f", positive=True)

# Conservation of momentum
# Because m_2 is at rest the equation simplifies to: m_1 vi = (m1 + m2) 1/3 vi
eq_momentum = sp.Eq(
    m_1 * v_i,
    (m_1 + m_2) * sp.Rational(1, 3) * v_i,
)
solution = sp.solve(eq_momentum, (m_2))

# 1. Extract the expression from the list
m_2_formula = solution[0]

# 2. Print it normally
print("Formula for m_2:")
m_2_formula
```

```python
# -----------------------------------------------------------------
# 2. Numerical Evaluation with AstroPy Units
#
# Perfectly inelastic collision: m_1 v_{1,i} + m_2 v_{2,i} =  (m_1 + m_2) v_{f}
# -----------------------------------------------------------------

# Given parameters with AstroPy units
mass_1 = 3 * u.kg

# conservation of momentum:  m_1 v_{1,i} + m_2 v_{2,i} =  (m_1 + m_2) v_{f}
# We apply the previous obtained formula
mass_2 = 2 * mass_1
print("--- AstroPy Numerical Result ---")
print(f"Mass of 2nd sphere: {mass_2:.0f}")
```

#### Problem 3 

A 10 g object is moving to the right at 20 cm/s and makes an head on elastic collision with a 15 g object. The 15 g object is initially moving in the opposite direction at 30 cm/s. What is the final velocity of each object after the collision?

<img src="../pics/Phy1_video_13_2.png" width=800> 


```python
# ==============================================================================
# STEP 1: SYMBOLIC FORMULAS
# ==============================================================================

# Define symbols for the algebraic equations
m_1, m_2, v_1i, v_2i, v_1f, v_2f = sp.symbols(
    "m_1, m_2, v_1i, v_2i, v_1f, v_2f", positive=True
)

# Explicit 1D elastic collision formulas matching the physical definition
v1f_formula = ((m_1 - m_2) * v_1i + 2 * m_2 * v_2i) / (m_1 + m_2)
v2f_formula = (2 * m_1 * v_1i + (m_2 - m_1) * v_2i) / (m_1 + m_2)

# Create lambdified functions.
# CRITICAL: Passing [m_1, m_2, v_1i, v_2i] as a list of symbols forces the exact positional order!
v1f_func = sp.lambdify([m_1, m_2, v_1i, v_2i], v1f_formula)
v2f_func = sp.lambdify([m_1, m_2, v_1i, v_2i], v2f_formula)

print("--- SymPy Symbolic Formulas Generated ---")
display(sp.Eq(v_1f, v1f_formula))
display(sp.Eq(v_2f, v2f_formula))
```

```python
# ==============================================================================
# STEP 2: PASS NUMERICAL QUANTITIES DIRECTLY TO THE FUNCTIONS
# ==============================================================================

# Define your parameters with AstroPy units
mass1 = 10 * u.g
vel1i = 20 * (u.cm / u.s)

mass2 = 15 * u.g
vel2i = -30 * (u.cm / u.s)  # Negative because it moves left

# Call the functions directly by passing the variables in the exact order: (m_1, m_2, v_1i, v_2i)
final_v1 = v1f_func(mass1, mass2, vel1i, vel2i)
final_v2 = v2f_func(mass1, mass2, vel1i, vel2i)

# --- DISPLAY RESULTS ---
print("--- Final Velocities (Original Units) ---")
print(f"v_1f = {final_v1:.2f}")
print(f"v_2f = {final_v2:.2f}")

print("\n--- Final Velocities Converted to SI (m/s) ---")
print(f"v_1f = {final_v1.to(u.m / u.s):.1f}")
print(f"v_2f = {final_v2.to(u.m / u.s):.1f}")
```
