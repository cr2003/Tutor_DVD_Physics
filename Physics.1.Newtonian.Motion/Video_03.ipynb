#### Helper functions

```python
def fmt(val, decimals=2):
    """Devuelve un entero si no tiene decimales, si tiene decimales los redondea a la cantidad pedida en decimals"""
    n = float(val)
    n = round(n, decimals)
    if n == int(n):
        return f"{int(n):,}"
    else:
        return f"{n:,.{decimals}f}"
```

#### Problem

Two people pull on a mule, as seen from above.

a) What is the single force that is equivalent to the two forces given?

b) What force would a third person have to exert to make the net force on the mule equal to zero?

<img src="../pics/Phy1_video_3_2.png" width=600> 


```python
### a)
F1 = 120  # u.N
F2 = 80  # u.N

# Angles
theta1 = sp.rad(60)  # convert to radians
theta2 = sp.rad(180 - 75)  # convert to radians

# Vectors
F1_vec = sp.Matrix([F1 * sp.cos(theta1), F1 * sp.sin(theta1)])

F2_vec = sp.Matrix([F2 * sp.cos(theta2), F2 * sp.sin(theta2)])

# Resultant
Fr = F1_vec + F2_vec
Frx, Fry = Fr

# Magnitude
Fr_mag = sp.sqrt(Frx**2 + Fry**2)

# Direction
Fr_angle = sp.deg(atan2(Fry, Frx))
Fr_direction = sp.N(Fr_angle)


print(f"Result force: {sp.N(Fr_mag):.0f} N, direction: {Fr_direction:.0f}°")
```

```python
### b)

# Equilibrant F3

F3 = -Fr
print(f"Equilibrant force: {sp.N(Fr_mag):.0f} N, direction: {Fr_direction + 180:.0f}°")
```
