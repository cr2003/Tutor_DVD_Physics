##  Work

$$W = F \cdot d $$

F, d are in the same direction. Unit is joule J. 




#### Problem 1

A tugboat pulls a large ship with a force of $5\times 10^3$ N. How much work does the butboat do on the ship, if they both move a distance of 3 km?

<img src="../pics/Phy1_video_7.png" width=600> 


```python
# -----------------------------------------------------------------
# Numerical Evaluation with AstroPy Units
# -----------------------------------------------------------------

F = 5e3 * u.N
d = 3 * u.km
W = F * d
print(f"Work: {W.to(u.J):.1e} ≈ {W.to(u.kJ):,.1e}")
```

#### Problem 2

A person pushes a shopping cart with a force of 35 N directed downward at an angle os 25 deg from the horizontal. If the person pushes the cart 50 m, how much work is done on the cart?

<img src="../pics/Phy1_video_7_2.png" width=600> 


```python
# -----------------------------------------------------------------
# Numerical Evaluation with AstroPy Units
# -----------------------------------------------------------------

F = 35 * u.N
theta = 25 * u.deg
d = 50 * u.m
W = F * np.cos(theta) * d
print(f"Work: {W.to(u.J):,.0f} ≈ {W.to(u.kJ):,.1f}")
```
