##  Power


$$Power = \frac{Work}{Time} $$

Units $$\left[\frac{J}{s} = W \right] $$

$$1 HP = 746 W$$

$$P =\frac{W}{\Delta t} = \frac{F \cdot d}{\Delta t} = F \cdot \frac{d}{\Delta t} = F \cdot v$$

$$P = F \cdot v$$



#### Problem 1

A car has a mass of $1.5\times10^3$ kg and is accelerated from rest to 10m/s in 3 s. What is:

a) the work done on the car during this time?

<img src="../pics/Phy1_video_10.png" width=800> 



```python
# -----------------------------------------------------------------
# 2. Numerical Evaluation with AstroPy Units
# -----------------------------------------------------------------

# Given parameters with AstroPy units
mass = 1.5e3 * u.kg
vel_i = 0 * (u.m / u.s)
vel_f = 10 * (u.m / u.s)
delta_t = 3 * u.s


# Work: W = F d
# F = m a
# a = (v_f - v_i) / delta_t
# d = v0 t + 1/2 a t^2
accel = (vel_f - vel_i) / delta_t
F = mass * accel
d = vel_i * delta_t + 1 / 2 * accel * delta_t**2
W = F * d

print("--- AstroPy Numerical Result ---")
print(f"Work: {W.to(u.J):,.0f} ≈ {W.to(u.kJ):,.0f}")
```

b) the average power delivered to the car by the engine?

```python
P = W / delta_t
P.to(u.W)
print(f"Average Power: {P.to(u.W):,.0f} ≈ {P.to(u.imperial.hp):,.1f}")
```

c) the instantaneous power delivered to the car by the engine at t=2?

```python
# We calculate the velocity at 2 s
t_2 = 2 * u.s
vel_2 = vel_i + accel * t_2

# Formula for instantaneous Power
P = F * vel_2
print(f"Instantaneous Power at 2 s.: {P.to(u.W):,.0f} ≈ {P.to(u.imperial.hp):,.1f}")
```
