#### Average velocity
Units in meters per second

$$\bar{v} = \frac{\Delta x}{\Delta t} = \frac{x_f - x_i}{t_f - t_i} = \left[\frac{m}{s}\right] $$

#### Acceleration

$$\bar{a} = \frac{\Delta v}{\Delta t} = \frac{v_f - v_i}{t_f - t_i} = \left[\frac{m/s}{s} \right] =  \left[\frac{m}{s^2}\right]$$


##  Equations of Motion in 1 Dimension

### Motion with constant acceleration

$$ x = v_0 t + \frac{1}{2}at^2 $$

$$v = v_0 + at$$

$$v^2 = v_0^2 + 2ax$$

##  Projectile Motion

### x-dir

$$v_x = v_{x0} = v_0 \cos \theta$$

$$x = v_{x0} t = (v_0 \cos \theta) t$$

### y-dir

$$v_y = v_{y0} -gt $$

$$y = v_{y0} t - \frac{1}{2} g t^2$$

$$v_{y}^2 = v_{y0}^2 -2gy $$




### 2. The Second Law: The Law of Force and Acceleration

The acceleration of an object depends directly upon the net force acting upon the object and inversely upon the mass of the object.

$$F = m a $$

Newton Units:
$$(kg)(\frac{m}{s^2}) = N$$

##  Static Friction

$$f_s \leq \mu_s F_N$$

##  Kinetic Friction

$$f_k = \mu_k F_N$$

#### Coefficients 

$$\mu_k < \mu_s$$


##  Work

$$W = F \cdot d $$

F, d are in the same direction. Unit is joule J. 




##  Kinetic Energy

$$KE = \frac{1}{2}m v^2 $$

Units of Jules [J]

### Work - Energy Theorem

$$W_{net} = KE_f - KE_i = \Delta KE$$

$$W_{net} = \frac{1}{2}m v_f^2 - \frac{1}{2}m v_i^2$$





##  Potentiall Energy

$$PE = mgh $$

Units of Jules [J]


##  Potentiall Energy


$$PE = \frac{1}{2}kx^2 $$






##  Mechanical Energy

$$ME = KE + PE$$

Conserved or Constant

$$ KE_i + PE_i = KE_f + PE_f $$

$$ \frac{1}{2}mv_i^2 + mgh_i = \frac{1}{2}mv_f^2 + mgh_f $$



##  Power


$$Power = \frac{Work}{Time} $$

Units $$\left[\frac{J}{s} = W \right] $$

$$1 HP = 746 W$$

$$P =\frac{W}{\Delta t} = \frac{F \cdot d}{\Delta t} = F \cdot \frac{d}{\Delta t} = F \cdot v$$

$$P = F \cdot v$$



##  Momentum

It's a vector defined by mass times velocity

$$\vec{p}= m \vec{v} $$

Also it's related to Force

$$ F = \frac{\Delta p}{\Delta t}$$

Impulse ($F \Delta t$)

$$F \Delta t = m v_f - m v_i$$

##  Conservation of Momentum

$$p_i = p_f$$

$$m1 \; v1_i + m2 \;  v2_i = m1 \;  v1_f + m2 \; v2_f $$



##  Elastic Collision

P & KE Conserved  (billiard balls)

Conservation of Momentum:

$$m_1 v_{1,i} + m_2 v_{2,i} =  m_1 v_{1,f} + m_2 v_{2,f}$$

Conservation of KE:

$$\frac{1}{2}m_1 v_{1,i}^2 + \frac{1}{2}m_2 v_{2,i}^2 =  \frac{1}{2}m_1 v_{1,f}^2 + \frac{1}{2}m_2 v_{2,f}^2$$


##  Inelastic Collision

Conservation of Momentum:

$$m_1 v_{1,i} + m_2 v_{2,i} =  m_1 v_{1,f} + m_2 v_{2,f}$$

##  Perfectly Inelastic Collision
The two object in the collision stick together resulting only one final compound object

Conservation of Momentum:

$$m_1 v_{1,i} + m_2 v_{2,i} =  (m_1 + m_2)  v_f$$

##  Angular Speed and Angular Acceleration

$$360^{\circ} = 2\pi rad$$

### Arc Length

$$s = r \theta$$

### Angular Speed

$$\bar{\omega} = \frac{\theta_2 - \theta_1}{t_2 - t_1} = \frac{\Delta \theta }{\Delta t} \implies Units [\frac{rad}{s}]$$

### Angular Acceleration

$$\bar{\alpha} = \frac{\omega_2 - \omega_1}{t_2 - t_1} = \frac{\Delta \omega }{\Delta t} \implies Units [\frac{rad}{s^2}]$$


###  Rotational Equations of Motion

$$ \omega = \omega_0 + \alpha t  $$

$$ \theta  = \omega_0 t+ \frac{1}{2}\alpha t^2  $$

$$ \omega^2 = \omega_0^2 + 2 \alpha \theta  $$



##  Tangential Velocity

$$ v_t = r \omega  $$


##  Tangential Acceleration

$$ a_t = r \alpha  $$



## Centripetal Acceleration

$$ a_c = \frac{v_t^2}{r}  $$



## Centripetal Force

$$ a_c = \frac{v_t^2}{r}\;\;, \;\; F = m a  $$

$$F_c = m a_c =  \frac{m v_t^2}{r} $$




##  Gravitational Force

$$ F = \frac {G m_1 m_2}{r^2} $$

$$ G = 6.673 \times 10^{-11} \; \frac{N m^2}{kg^2}$$


###  Kepler's Laws of Orbital Motion

1) All planets move in elliptical orbits with respect to the sun at one foci

2) A line draw from the Sun to the planet sweeps out equal areas in equal times


3) The orbital period is given by:

$$ T^2 = \left( \frac{4\pi^2}{G M_s} \right) r^3$$
