##  The Buoyant Force

<img src="../pics/Phy1_video_23.png" width=400> 

In a pool the pressure increases as you go down in the pool.

The pressure equals to the weight of the column of water above you.

$$ P = \rho g h$$

$$ P_{Total} = P_0 + \rho g h$$

Normally, $P_0$ is the atmospheric pressure.



<img src="../pics/Phy1_video_23_2.png" width=400> 

The buoyant force is the weight of the liquid displaced by the object.

$$F_B = V \rho g$$

#### Problem 1

A frog is in a circular salad bowl which is floating in a liquid with density 1.35 g/cm^3. If the salad bowl has radius 6cm and negligiblel mass, what is the mass of the frog?

<img src="../pics/Phy1_video_23_3.png" width=400> 


```python
# Define known values with units
r = 6 * u.cm
rho = 1.35 * (u.g / u.cm**3)


# calculate the volumen of the salad bowl: V = (1/2)(4/3) * pi * r^3
V = (1 / 2) * (4 / 3) * np.pi * r**3

# calculate the buoyantforce: F_B = V * rho * g
F_B = V * rho * const.g0

# calculate the mass of the frog
# mass frog + mass salad = F_B / g; as mass of salad is negligiblel ==> m_frog = F_B / g
mass_frog = F_B / const.g0

print(f"Frog mass: {mass_frog.to(u.kg):.2f}")
```

#### Problem 2

A beaker is filled with water. A spring of constant k=16 N/m is connected to the bottom of the beaker and a 5 kg block of wood is connected to the top of the spring. How much does the spring stretch?

<img src="../pics/Phy1_video_23_4.png" width=600> 



```python
# Define known values with units
k = 16 * (u.N / u.m)
mass_wood = 5 * u.kg
rho_wood = 650 * (u.kg / u.m**3)
rho_water = 1e3 * (u.kg / u.m**3)

# Calculate the volume of the wood block: rho = m / V ==> V = m / rho
V_wood = mass_wood / rho_wood
# Calculate the weight of the wood W = m * g
W_wood = mass_wood * const.g0

# calculate the buoyant force: F_B = V * rho * g
F_B = V_wood * rho_water * const.g0


# Hooke's Law: F = -kx ==> F_spring = k * delta_L


# Equilibrium: F_B = W_wood + F_spring ==> F_B = W_wood + k * delta_L ==> delta_L = (F_B - W_wood) / k
delta_l = (F_B - W_wood) / k

print(f"Spring stretch: {delta_l.to(u.m):.2f}")
```
