```python
def fmt(val, decimals=2):
    """Devuelve un entero si no tiene decimales, si tiene decimales los redondea a la cantidad pedida en decimals"""
    n = float(val)
    n = round(n, decimals)
    if n == int(n):
        return f"{int(n):,}"
    else:
        return f"{n:,.{decimals}f}"
```

#### Problem 1

A copper block has a mass of 2 kg and rests on a steel table. It is connected by a string to a mass M and the wall as shown. If the angle from the string to the wall is 30 degrees, what is the maximum value of M such that the smaller block does not move?

<img src="../pics/Phy1_video_6.png" width=600> 


```python
# -----------------------------------------------------------------
# 1. Symbolic Solution with SymPy
# -----------------------------------------------------------------

# Define symbols for the algebraic solution
# m: mass of small block, M: mass of suspended block, mu_s: static friction coeff
# T: tension of angled string, T_2: tension of horizontal string
# theta: string angle (30 degrees)
m, M, mu_s, g, T, T_2, theta = sp.symbols("m M mu_s g T T_2 theta", positive=True)

# Equation 1: Maximum static friction balances horizontal tension T_2
# F_N = m * g -> f_s_max = mu_s * m * g
eq_block = sp.Eq(T_2, mu_s * m * g)

# Equation 2: Horizontal balance at the knot (T * cos(30 deg) = T_2)
eq_knot_x = sp.Eq(T * sp.cos(theta), T_2)

# Equation 3: Vertical balance at the knot (T * sin(30 deg) = M * g)
eq_knot_y = sp.Eq(T * sp.sin(theta), M * g)

# Solve the system of equations for T, T_2, and M
solution = sp.solve([eq_block, eq_knot_x, eq_knot_y], (T, T_2, M))

# Extract the symbolic formula for mass M
M_formula = solution[M]
print("--- SymPy Symbolic Formula ---")
print(f"M = {M_formula}\n")
```

```python
solution[M]
```

```python
# -----------------------------------------------------------------
# 2. Numerical Evaluation with AstroPy Units
# -----------------------------------------------------------------

# Given parameters with AstroPy units
mass_small_block = 2.0 * u.kg
coefficient_static_friction = 0.53
theta_val = 30 * u.deg

# Convert the SymPy expression to a callable function
M_func = sp.lambdify((m, mu_s, theta), M_formula)

# Calculate numerical value (g cancels out in the formula, but AstroPy handles the units)
M_numeric_value = M_func(mass_small_block, coefficient_static_friction, theta_val)


print("--- AstroPy Numerical Result ---")
print(f"Maximum Mass M: {M_numeric_value:.3f}")
```

#### Problem 2

A box wighinng 300 N is pushed across the floor with a force of 400 N directed down at an angle of 35.2 degrees. The coefficient of kinetic friction is 0.570. How long does it take for the box to move 4 m?

<img src="../pics/Phy1_video_6_2.png" width=600> 


```python
import numpy as np

# Define given parameters with explicit AstroPy units
weight_box = 300.0 * u.N
F = 400.0 * u.N
theta = 35.2 * u.deg
mu_k = 0.570  # kinetic friction
x = 4.0 * u.m  # displacement
g = const.g0  # gravity

# Step 1: Deconstruct the applied force
# Numpy knows how to apply astropy deg to rad
Fx = F * np.cos(theta)
Fy = F * np.sin(theta)

# Step 2: Determine the normal force F_N and frictiong f_k
# Vertical equilibrium (ΣF_y = 0) ==> F_N - W - Fy = 0 ==> F_N = W + Fy
F_N = weight_box + Fy

# kinetic friction force f_k
# f_k = mu_k F_N
f_k = mu_k * F_N

# Step 3: Find the acceleration
m = weight_box / g

# ΣF_x = Fx - f_k = ma ==> a = Fx - f_k / m
a = (Fx - f_k) / m

# Step 4: calculate the time t to move d (4 m)
# kinematic displacement equation x = 1/2 a t^2 (initial velocity 0) ==> t = sqrt(2x / a)
t = np.sqrt((2 * x) / a)

print(f"time for box to move 4 m: {t:.1f}")
```
