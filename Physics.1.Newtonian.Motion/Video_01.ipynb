#### Helper functions

```python
def fmt(val, decimals=2):
    """Devuelve un entero si no tiene decimales, si tiene decimales los redondea a la cantidad pedida en decimals"""
    n = float(val)
    n = round(n, decimals)
    if n == int(n):
        return f"{int(n):,}"
    else:
        return f"{n:,.{decimals}f}"
```

###  Velocity and Acceleration in One Dimension

#### Displacement
The movement from initial to final position

$$\Delta x = x_f - x_i$$

Units in meters

$$\Delta x = [m]$$

#### Velocity
Units in meters per second

$$velocity = [m/s] $$

#### Average velocity
Units in meters per second

$$\bar{v} = \frac{\Delta x}{\Delta t} = \frac{x_f - x_i}{t_f - t_i} = \left[\frac{m}{s}\right] $$

#### Problem
A person runs a marathon in 2h, 9 min, 21 s. The marathon distance is 26 mi, 385 yd.

Calculate the average speed in mile per hour.

```python
delta_t = 2 * u.h + 9 * u.min + 21 * u.s
delta_x = 26 * u.imperial.mi + 385 * u.imperial.yd

avg_v = delta_x / delta_t
print(f"Average speed: {avg_v:.1f}")
```

#### Acceleration

$$\bar{a} = \frac{\Delta v}{\Delta t} = \frac{v_f - v_i}{t_f - t_i} = \left[\frac{m/s}{s} \right] =  \left[\frac{m}{s^2}\right]$$


#### Problem
We are sending a man to the moon by shooting him out of a canon. The canon is 220 m long and velocity of the person as he exits the canon is 10.97 km/s. What acceleration does he experience as he travels down the canon?

<img src="../pics/Phy1_video_1.png" width=400> 


```python
delta_x = 220 * u.m
v_f = 10.97 * u.km / u.s
v_f = v_f.to(u.m / u.s)
v_i = 0
avg_v = (v_i + v_f) / 2

# v = delta_x / t ==> t = delta_x / v
t = delta_x / avg_v
avg_acc = (v_f - v_i) / t

print(f"Average acceleration: {avg_acc:,.0f}")
print(f"The g factor is: {avg_acc / const.g0:,.0f} g")
```
