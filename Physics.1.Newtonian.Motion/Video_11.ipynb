##  Momentum

It's a vector defined by mass times velocity

$$\vec{p}= m \vec{v} $$

Also it's related to Force

$$ F = \frac{\Delta p}{\Delta t}$$

Impulse ($F \Delta t$)

$$F \Delta t = m v_f - m v_i$$

#### Problem 1

A baseball with mass 0.145 kg is thrown with as much momentum as a 3 g bullet moving with speed of $1.5 \times 10^3$ m/s. How fast the ball be thrown for this to be true?

<img src="../pics/Phy1_video_11_2.png" width=800> 



```python
# -----------------------------------------------------------------
# 2. Numerical Evaluation with AstroPy Units
# -----------------------------------------------------------------

# Given parameters with AstroPy units
mass_ball = 0.145 * u.kg
mass_bullet = 3 * u.g
vel_bullet = 1.5e3 * (u.m / u.s)

# momentum:  mass_ball * vel_ball = mass_bullet * vel_bullet
vel_ball = mass_bullet * vel_bullet / mass_ball
print("--- AstroPy Numerical Result ---")
print(f"Ball velocity: {vel_ball.to(u.m / u.s):.0f}")
```

#### Problem 2

A garden hose is held parallel to the ground. Water is flowing out of the hose at a rate of 0.6 kg/s with a speed of 25 m/s. What force is needed to hold the hose stationary?

<img src="../pics/Phy1_video_11.png" width=800> 



```python
# -----------------------------------------------------------------
# Numerical Evaluation with AstroPy Units
# -----------------------------------------------------------------

# Given parameters with AstroPy units
vel = 25 * (u.m / u.s)
delta_mass_time = 0.6 * (u.kg / u.s)

# F = delta p / delta t = (delta m / delta t) * v

F = delta_mass_time * vel

print(f"Force: {F.to(u.N):.0f}")
```

#### Problem 3

A 3 kg ball hits a wall at 10 m/s at an angle of 60 degrees. It bounces off of the wall at the same speed and angle. If the contact time between the ball and wall is 0.20 sec, what is the average force exerted on the ball by the wall?

<img src="../pics/Phy1_video_11_2.png" width=600> 



```python
# -----------------------------------------------------------------
# Numerical Evaluation with AstroPy Units
# -----------------------------------------------------------------

# Given parameters with AstroPy units
mass = 3 * u.kg
vel = 10 * (u.m / u.s)
theta = 60 * u.deg  # (related to the wall, axis-y)
delta_t = 0.2 * u.s

# Change in velocity related to axis-x (bounce of the ball)
v_xi = vel * np.sin(theta)

# bounce in the wall, change to the opposite side
v_xf = -v_xi

# Impulse: F delta_t = m(v_xf - v_xi) ==> F = m(v_xf - v_xi) / delta_t
F = (mass * (v_xf - v_xi)) / delta_t

print(f"Force: {F.to(u.N):.0f}")
print("The -sign indicates the force is to the left, negative direction in the x-axis")
```
