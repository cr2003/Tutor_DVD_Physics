### Angular Speed

$$\bar{\omega} = \frac{\theta_2 - \theta_1}{t_2 - t_1} = \frac{\Delta \theta }{\Delta t} \implies Units [\frac{rad}{s}]$$

### Angular Acceleration

$$\bar{\alpha} = \frac{\omega_2 - \omega_1}{t_2 - t_1} = \frac{\Delta \omega }{\Delta t} \implies Units [\frac{rad}{s^2}]$$

#### Problem 1

Convert 30 deg and 270 deg to radians. 


```python
angle = 30 * u.deg

print(f"30 deg is {angle.to(u.rad):.2f}")
```

```python
angle = np.pi / 2 * u.rad

print(f"Pi/2 rad is {angle.to(u.deg):.0f}")
```

#### Problem 2

If a wheel has a radius of 4.1 m and the wheel rotates through an angle of 30 deg, how far along its circumference does it travel?


```python
theta = 30 * u.deg
theta = theta.to(u.rad)
r = 4.1 * u.m

s = r * theta.value
print(f"Travel: {s:.2f}")
```

#### Problem 3

A wheel moves from rest to an angular speed of 0.20 revolutions/second in 30 seconds. What is its angular acceleration?



```python
omega_i = 0 * (u.rad / u.s)
omega_f = 0.20 * (u.cycle / u.s)
delta_t = 30 * u.s

alpha = (omega_f - omega_i) / delta_t
alpha_rad = alpha.to(u.rad / u.s**2)
print(f"alpha: {alpha_rad:.2e}")
```
