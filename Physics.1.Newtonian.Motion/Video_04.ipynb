#### Helper functions

```python
def fmt(val, decimals=2):
    """Devuelve un entero si no tiene decimales, si tiene decimales los redondea a la cantidad pedida en decimals"""
    n = float(val)
    n = round(n, decimals)
    if n == int(n):
        return f"{int(n):,}"
    else:
        return f"{n:,.{decimals}f}"
```

##  Projectile Motion

### x-dir

$$v_x = v_{x0} = v_0 \cos \theta$$

$$x = v_{x0} t = (v_0 \cos \theta) t$$

### y-dir

$$v_y = v_{y0} -gt $$

$$y = v_{y0} t - \frac{1}{2} g t^2$$

$$v_{y}^2 = v_{y0}^2 -2gy $$

```python
# =====================================================================
# 1. SYMBOLIC STAGE (SymPy) - Calculate t to reach the goal
# =====================================================================

# Define abstract mathematical symbols
(
    x,
    v0,
    theta,
    t,
    a,
) = sp.symbols("x, v0, theta, t, a", positive=True)

# Define the kinematic equation: x = v_0 * cos(theta) * t
equation = sp.Eq(x, v0 * sp.cos(theta) * t)

# Solve the equation for the unknown time 't'
solutions = sp.solve(equation, t)
t_isolated = solutions[0]

print("--- time Formula Solved by SymPy ---")
display(Math(rf"{sp.latex(sp.Eq(t, t_isolated))}"))
```

#### Problem

You throw a ball straight upward and it returns to your hand after 3 s in the air. A second ball is thrown at an angle of 30 degrees. What speed must the second ball be thrown so that it reaches the same height as the first ball?

<img src="../pics/Phy1_video_4_2.png" width=600> 


```python
# =====================================================================
# 1. UNITS AND NUMERICAL STAGE (Pure AstroPy + NumPy)
# The first ball its up in the air 3s. 1.5 up and 1.5 down.
# In the down, it starts at vy0 = 0, so we calculate how much its
# =====================================================================

# 1. Define your inputs with AstroPy units
t_down = 1.5 * u.s
a = -const.g0  # gravity
# kinematic equation: y = 0 + 1/2 * a * t^2
y_result = 0.5 * a * t_down**2

print("--- a) y result ---")
print(f"High reach by ball: {y_result:.2f}")
print(f"The - sign indicates the down fall, the height is {-y_result:.2f}")
```

```python
# =====================================================================
# 2. SYMBOLIC STAGE (SymPy) - Calculate v0
# =====================================================================

# Define abstract mathematical symbols
x, y, v0, theta, a, zero = sp.symbols("x, y, v0, theta, a, zero", positive=True)

# Define the kinematic equation: 0 = (v_0 * sin(theta))**2 + 2ay
equation = sp.Eq(zero, (v0 * sp.sin(theta)) ** 2 + 2 * a * y)

# Solve the equation for the unknown v0
solutions = sp.solve(equation, v0)
v0_isolated = solutions[1]

print("--- v0 Formula Solved by SymPy ---")
display(Math(rf"{sp.latex(sp.Eq(v0, v0_isolated))}"))
```

```python
# =====================================================================
# 3. UNITS AND NUMERICAL STAGE (Pure AstroPy + NumPy)
# The second ball its thrown with an initial velocity v0 and in 1.5 s. reaches
# its max high.
# =====================================================================

# 1. Define your inputs with AstroPy units
a = -const.g0  # gravity
y_high = -y_result
theta_val = 30 * u.deg
# kinematic equation: v0 = -np.sqrt(-2*a*y) / sin(theta)
v0_result = np.sqrt(-2 * a * y_high) / np.sin(theta_val)

print("--- a) v0 result ---")
v0_result
print(f"The second ball initial velocity: {v0_result:.1f}")
```
