#### Problem 1

A person with mass 50 kg runs at 10 m/s and jumps over a bar. If the horizontal component of her motion is 1 m/s when she is over the bar, how high did she jump?

<img src="../pics/Phy1_video_9_3.png" width=600> 


```python
# -----------------------------------------------------------------
# 1. Symbolic Solution with SymPy
# -----------------------------------------------------------------

# Define symbols for the algebraic solution
m, v_i, v_f, h_i, h_f, g = sp.symbols("m, v_i, v_f, h_i, h_f, g", positive=True)
# Conservation of Energy Equation
eq_const = sp.Eq(
    sp.Rational(1, 2) * m * v_i**2 + m * g * h_i,
    sp.Rational(1, 2) * m * v_f**2 + m * g * h_f,
)
solution = sp.solve(eq_const, (h_f))

# 1. Extract the expression from the list
h_formula = solution[0]

# 2. Print it normally
print("Formula for h_f:")
h_formula
```

```python
# -----------------------------------------------------------------
# 2. Numerical Evaluation with AstroPy Units
# -----------------------------------------------------------------

# Given parameters with AstroPy units
mass = 50 * u.kg
vel_i = 10 * (u.m / u.s)
vel_f = 1 * (u.m / u.s)
gravity = const.g0
high_i = 0 * u.m  # Person runs over the floor

# Convert the SymPy expression to a callable function
h_func = sp.lambdify((m, g, h_i, v_i, v_f), h_formula)

# Calculate numerical value
h_numeric_value = h_func(mass, gravity, high_i, vel_i, vel_f)


print("--- AstroPy Numerical Result ---")
print(f"high: {h_numeric_value:.1f}")
```

#### Problem 2

Tarzan swings on a 30 m long rope that is initially inclined at 37 deg to the vertical. What is his speed at the bottom of the swing if:

a) he starts from rest?

<img src="../pics/Phy1_video_9_4.png" width=600> 


```python
# From the conservation of Energy
# KEi + PEi = KEf * PEf
# Initially KEi = 0 as vi=0, and at the bottom PEf = 0 as hf = 0
# The resultant eq: mghi = 1/2 m vf^2

# -----------------------------------------------------------------
# 1. Symbolic Solution with SymPy
# -----------------------------------------------------------------

# Define symbols for the algebraic solution
m, v_i, v_f, h_i, h_f, g = sp.symbols("m, v_i, v_f, h_i, h_f, g", positive=True)
# Conservation of Energy Equation
eq_const = sp.Eq(m * g * h_i, sp.Rational(1, 2) * m * v_f**2)
solution = sp.solve(eq_const, (v_f))

# 1. Extract the expression from the list
vf_formula = solution[0]

# 2. Print it normally
print("Formula for v_f:")
vf_formula
```

```python
# -----------------------------------------------------------------
# 2. Numerical Evaluation with AstroPy Units
# -----------------------------------------------------------------

# Given parameters with AstroPy units
theta = 37 * u.deg
gravity = const.g0
long = 30 * u.m  # length of the rope
high_i = long - long * np.cos(theta)

# Convert the SymPy expression to a callable function
vf_func = sp.lambdify((g, h_i), vf_formula)

# Calculate numerical value
vf_numeric_value = vf_func(gravity, high_i)


print("--- AstroPy Numerical Result ---")
print(f"vf: {vf_numeric_value:.1f}")
```

b) he pushes of with a speed of 4 m/s?

```python
# From the conservation of Energy
# KEi + PEi = KEf * PEf
# now we have to take care of the initial velocity
# The resultant eq: 1/2 m vi^2 + mghi = 1/2 m vf^2

# -----------------------------------------------------------------
# 1. Symbolic Solution with SymPy
# -----------------------------------------------------------------

# Define symbols for the algebraic solution
m, v_i, v_f, h_i, h_f, g = sp.symbols("m, v_i, v_f, h_i, h_f, g", positive=True)
# Conservation of Energy Equation
eq_const = sp.Eq(
    sp.Rational(1, 2) * m * v_i**2 + m * g * h_i, sp.Rational(1, 2) * m * v_f**2
)
solution = sp.solve(eq_const, (v_f))

# 1. Extract the expression from the list
vf_formula = solution[0]

# 2. Print it normally
print("Formula for v_f:")
vf_formula
```

```python
# -----------------------------------------------------------------
# 2. Numerical Evaluation with AstroPy Units
# -----------------------------------------------------------------

# Given parameters with AstroPy units
vel_i = 4 * (u.m / u.s)

# Convert the SymPy expression to a callable function
vf_func = sp.lambdify((g, h_i, v_i), vf_formula)

# Calculate numerical value
vf_numeric_value = vf_func(gravity, high_i, vel_i)


print("--- AstroPy Numerical Result ---")
print(f"vf: {vf_numeric_value:.1f}")
```
