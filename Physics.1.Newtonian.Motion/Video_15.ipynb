##  Rotational Equations of Motion

$$ \omega = \omega_0 + \alpha t  $$

$$ \theta  = \omega_0 t+ \frac{1}{2}\alpha t^2  $$

$$ \omega^2 = \omega_0^2 + 2 \alpha \theta  $$



#### Problem 1

A dentist's drill starts from rest and accelerates to a final angular velocity of $2.51 \times 10^4$ rev/min over a period of 3.2 seconds. What is:

a) The angular acceleration?


```python
omega_i = 0 * (u.rad / u.s)
omega_f = 2.51e4 * (u.cycle / u.min)
delta_t = 3.2 * u.s

alpha = (omega_f - omega_i) / delta_t
alpha_rad = alpha.to(u.rad / u.s**2)
print(f"alpha: {alpha_rad:.0f}")
```

b) The angle through which the drill rotates?

```python
theta = omega_i * delta_t + (1 / 2 * alpha_rad * delta_t**2)
print(f"The total angle: {theta:.0f}")
```

#### Problem 2

A whashing machine starts from rest and accelerates to an angular speed of 5 rev/s in 8 seconds. After this, the washing machine is unplugged and allowed to come to rest over a period of 12 sconds. Through how many revolutions does the tub turn?


```python
omega_i = 0 * (u.rad / u.s)
omega_f1 = 5 * (u.cycle / u.s)
omega_f1_rad = omega_f1.to(u.rad / u.s)
delta_t1 = 8 * u.s
delta_t2 = 12 * u.s


alpha_1 = (omega_f1_rad - omega_i) / delta_t1
theta_1 = (omega_i * delta_t1) + 1 / 2 * alpha_1 * delta_t1**2

alpha_2 = (omega_i - omega_f1_rad) / delta_t2
theta_2 = (omega_f1_rad * delta_t2) + 1 / 2 * alpha_2 * delta_t2**2

total_theta = theta_1 + theta_2
total_theta.to(u.cycle)
```

#### Problem 2

A car is driving at 30 m/s and hits the brakes which causes the car to undergo an acceleration of $-2m/s^2$. How many revolutions does each tire make before coming to a rest if the tires have a radii of 0.3 meters?


```python
vi = 30 * (u.m / u.s)
vf = 0 * (u.m / u.s)
acc = -2 * (u.m / u.s**2)
radius = 0.3 * u.m

# vf**2 = vi**2 + 2ax ==> x = (vf**2 - vi**2) / (2a)
x = (vf**2 - vi**2) / (2 * acc)
circ = 2 * np.pi * radius
rev = x / circ
print(f"Revolutions: {rev:.0f}")
```
