#### Problem 1

A race car goes around a circular track that has a radius of 400 meters. The car undergoes an acceleration of 0.5 m/s^2. At the point where the centripetal and tangetial accelerations are equal, find:

a) the speed of the car?


```python
r = 400 * u.m
a_t = 0.5 * (u.m / u.s**2)

# a_t = r alpha ==> alpha = a_t / r
alpha = a_t / r

# a_t = a_c;  r alpha = v_t^2 / r ==> v_t = sqrt(r^2 alpha)

v_t = np.sqrt(r**2 * alpha)

print(f"The speed of the car: {v_t:.1f}")
```

b) the distance traveled by the car?

```python
v_0 = 0 * (u.m / u.s)

# v^2 = v0^2 + 2 a x ==> x = (v^2 - v0^2) / (2 a)
x = (v_t**2 - v_0**2) / (2 * a_t)
print(f"Distance traveled by car: {x:.0f}")
```

c) the elapsed time

```python
# x = v0 t + 1/2 a t^2, as initial velocity is zero ==>  t = sqrt((2 x)/ a)
t = np.sqrt((2 * x) / a_t)

print(f"Time: {t:.1f}")
```

#### Problem 2

A car with mass 2,000 kg goes around a circular turn with a radius of 20 meters. The coefficient of friction between the tires and the road is 0.70. How fast can the car go without skidding during the turn?

<img src="../pics/Phy1_video_16_3.png" width=400> 



#### Solution.

1. Centripetal force: $F_c = \frac{m \cdot v^2}{r}$

2. Maximum friction force: $F_f = \mu \cdot m \cdot g$

3. Setting them equal to solve for the maximum speed (v_t).


```python
# ==============================================================================
# STEP 1: SYMBOLIC FORMULA
# ==============================================================================

# Definte symbols
m, v_t, r, mu, g = sp.symbols("m, v_t, r, mu, g", positive=True)

# Define the forces
F_centripetal = m * v_t**2 / r
F_friction = mu * m * g

# Set equations equal: F_centripetal = F_friction
equation = sp.Eq(F_centripetal, F_friction)

# Solve explicitly for velocity 'v_t'
v_solutions = sp.solve(equation, v_t)
v_formula = v_solutions[0]

# Create an exexutable function
v_max_func = sp.lambdify([r, mu, g], v_formula)

v_formula
```

```python
# ==============================================================================
# STEP 2: PASS NUMERICAL QUANTITIES DIRECTLY
# ==============================================================================

# Defince parameters with AstroPy units
radius = 20 * u.m
coefficient_friction = 0.70
gravity = const.g0

# Call the function directly by passing the parameters in order: (r, mu, g)
max_velocity = v_max_func(radius, coefficient_friction, gravity)

print("--- Maximum Sage Speed Result ---")
print(f"v_max = {max_velocity:.2f} equivalent to {max_velocity.to(u.km / u.h):.0f}")
```
