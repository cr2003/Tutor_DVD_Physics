##  The Bernoulli Equation

### Ideal Fluid
* nonviscous

### Continuity Equation



<img src="../pics/Phy1_video_24.png" width=400> 

$$ \text{Mass in} = \text{Mass out}$$

$$\rho A_1 V_1 = \left[ \frac{kg}{\cancel{m^3}} \cancel{m^2} \frac{\cancel{m}}{s}  = \frac{kg}{s} \right] $$

$$\cancel{\rho} A_1 V_1 = \cancel{\rho}A_2 V_2 $$

#### Eqn of continuity

$$A_1 V_1 = A_2 V_2 $$


###  The Bernoulli Equation

<img src="../pics/Phy1_video_24_2.png" width=600> 

It's a type of Conservation of Total Energy

$$ P_1 + \frac{1}{2} \rho v_1^2 + \rho g Y_1 = P_2 + \frac{1}{2} \rho v_2^2 + \rho g Y_2$$

A particular example, when a pipe narrows, and the median height is constant:

<img src="../pics/Phy1_video_24_3.png" width=600> 


$$ P_1 + \frac{1}{2} \rho v_1^2 = P_2 + \frac{1}{2} \rho v_2^2 $$

$$v_2 > v_1 \implies P_2 < P_1$$


#### Problem 1

A liquid with density of $1.65 g/cm^3$ flows through a pipe that changes diameter. The first section of pipe has a cross sectional area of $10 cm^2$, the flow speed is $275 cm/s$, and the pressure is $1.2\times 10^5 Pa$ in that secton. The 2nd section of pipe has cross sectional area of $2.5 cm^2$.

What is:

a) the flow speed in the 2nd section?

b) the pressure in the 2nd section?

<img src="../pics/Phy1_video_24_4.png" width=600> 



```python
# Define known values with units
rho = 1.65 * (u.g / u.cm**3)
A_1 = 10 * u.cm**2
V_1 = 275 * (u.cm / u.s)
P_1 = 1.2e5 * u.Pa
A_2 = 2.5 * u.cm**2

# a)
# calculate velocity in 2nd section, from continuity eqn: A_1 V_1 = A_2 V_2 ==> V_2 = (A_1 V_1) / A_2
V_2 = (A_1 * V_1) / A_2

print(f"Flow speed 2nd section: {V_2.to(u.m / u.s):.0f}")

# b)
# calculate pressure in 2nd section, from Bernoulli eqn P_1 + 1/2 rho V_1^2 = P_2 + 1/2 rho V_2^2
P_2 = P_1 + ((1 / 2) * rho * V_1**2) - ((1 / 2) * rho * V_2**2)
P_2
print(f"pressure in 2nd section: {P_2:.2e}")
```

#### Problem 2

What is the lift produced by the wing given on the board?

<img src="../pics/Phy1_video_24_5.png" width=600> 



```python
# Define known values with units
rho = 1.29 * (u.kg / u.m**3)
v_1 = 300 * (u.m / u.s)
v_2 = 280 * (u.m / u.s)
A = 20 * u.m**2

# Calculate de difference of pressure: from Bernoulli eqn P_1 + 1/2 rho V_1^2 = P_2 + 1/2 rho V_2^2
# P_1 - P_2 = (1/2) * rho (v_2^2 - v_1^2)
delta_p = 1 / 2 * rho * (v_2**2 - v_1**2)
print(f"The difference pressure {delta_p.to(u.Pa)}, sign - implies P_2 > P_1")

# formula for Pressure: P = F / A ==> F = P * A
f = np.abs(delta_p) * A

print(f"The total lift force: {f.to(u.N):.1e}")
```
