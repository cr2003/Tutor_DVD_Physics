#### Problem 1

A satellite has a mass m and moves in a circular orbit around Earth with speed "v" and at a height of 1,000 km above the Earth's surface. If the Earth's radius is $6.38 \times 10^6$ m and the Earth's mass is $5.98 \times 10^{24}$ kg, what is the orbital speed of the satellite?

<img src="../pics/Phy1_video_17_4.png" width=400> 

```python
import sympy as sp

# ==============================================================================
# STEP 1: SYMBOLIC FORMULA
# ==============================================================================

# Define symbols
m, v, R_earth, h, M, G = sp.symbols("m v R_earth h M G", positive=True)

# The total orbital radius is the Earth's radius plus the altitude height
r = R_earth + h

# Define the forces
F_gravitational = G * M * m / r**2
F_centripetal = m * v**2 / r

# Set equations equal to solve for velocity 'v'
equation = sp.Eq(F_gravitational, F_centripetal)

# Solve explicitly for velocity
v_solutions = sp.solve(equation, v)
v_formula = v_solutions[0]  # Grab the physical positive root

# Create the lambdified function with explicit positional arguments
v_orbital_func = sp.lambdify([R_earth, h, M, G], v_formula)

print("--- SymPy Symbolic Formula Generated ---")
display(sp.Eq(sp.Symbol("v_orbital"), v_formula))
```

```python
# ==============================================================================
# STEP 2: PASS NUMERICAL QUANTITIES DIRECTLY
# ==============================================================================

# Define the problem constants and parameters with AstroPy units
radius_earth = 6.38e6 * u.m
height = (
    1000 * u.km
)  # AstroPy automatically handles the km-to-m conversion dynamically!
mass_earth = 5.98e24 * u.kg
G_constant = const.G

# Call the function directly by passing parameters in the exact order: (R_earth, h, M, G)
orbital_speed = v_orbital_func(radius_earth, height, mass_earth, G_constant)

# --- DISPLAY RESULTS ---
print("--- Orbital Speed Result ---")
print(f"v = {orbital_speed:.2f}")

# Optional: Convert to km/s for standard aerospace tracking
print(f"v in km/s = {orbital_speed.to(u.km / u.s):.2f}")
```
