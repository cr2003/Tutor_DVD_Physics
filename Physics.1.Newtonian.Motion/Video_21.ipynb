##  Rotational Kinetic Energy and Angular Momentum

Recall that in Linear Motion:

$$KE = \frac{1}{2}m v^2$$

In Rotational:

$$KE_r = \frac{1}{2}I \omega^2$$

Conservation of Energy, the sum of the translational Kinetic Energy, plus the Rotational Kinetic Energy, plus the gravitational Potencial Energy in the initial moment must be equal to the final moment:

$$(KE_t + KE_r + PE_g)_i  = (KE_t + KE_r + PE_g)_f$$




Recall that in Linear Motion:

$$p = m v$$

In Rotational, the angular momentum L:

$$L = I \omega$$

Recall that in Linear Motion:

$$F = \frac{\Delta p}{\Delta t}$$

In Rotational Motion:

$$\tau = \frac{\Delta L}{\Delta t}$$


#### Problem 1

The hour hand and minute hand on a large outdoor clock are 2.7 m and 4.5 m in length, respectively. The mass of the hour hand is 60 kg and the mass of the minute hand is 100 kg. What is the total rotational kinetic energy of these hands?

<img src="../pics/Phy1_video_21.png" width=400> 


```python
# Define known values with units
hour_hand_len = 2.7 * u.m
minute_hand_len = 4.5 * u.m
hour_hand_mass = 60 * u.kg
minute_hand_mass = 100 * u.kg
omega_hour_hand = ((2 * np.pi * u.rad) / (12 * u.hr)).to(u.rad / u.s)
omega_minute_hand = ((2 * np.pi * u.rad) / (1 * u.hr)).to(u.rad / u.s)


# Calculate Inertia for each hand, I = 1/3 * M * L^2
I_hour_hand = (1 / 3) * hour_hand_mass * hour_hand_len**2
I_minute_hand = (1 / 3) * minute_hand_mass * minute_hand_len**2

# Calculate the rotational KE , KE_r = 1/2 * I * omega**2
KE_r_hour_hand = (1 / 2) * I_hour_hand * omega_hour_hand**2
KE_r_minute_hand = (1 / 2) * I_minute_hand * omega_minute_hand**2
KE_r_total = KE_r_minute_hand + KE_r_hour_hand

print(
    f"Total rotational KE: {KE_r_total.to(u.J, equivalencies=u.dimensionless_angles()):.2e}"
)
```
