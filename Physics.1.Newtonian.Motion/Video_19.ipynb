#### Problem 1

A window washer stands on a scaffold that is hung from two ropes, one on each end. If the scaffold weights 200 N and is 3 m long, what is the tension in the ropes when the 700 N worker stands 1 m from the end?

<img src="../pics/Phy1_video_19.png" width=400> 


```python
import sympy as sp

# Define symbols
FA, FB = sp.symbols("F_A F_B")  # Tensions in the left and right ropes
W_s, W_w = sp.symbols("W_s W_w")  # Weight of scaffold and worker
L, x = sp.symbols("L x")  # Length of scaffold and distance of worker from left end

# 1. Torque equilibrium about the left end (A)
# Clockwise torques are negative, counterclockwise are positive
# The scaffold's weight acts at its center of mass (L/2)
torque_eq = sp.Eq(FB * L - W_w * x - W_s * (L / 2), 0)

# 2. Vertical force equilibrium
force_eq = sp.Eq(FA + FB - W_w - W_s, 0)

# Solve the system of equations
symbolic_solutions = sp.solve([torque_eq, force_eq], (FA, FB))

# Display the symbolic expressions
print("Symbolic Solutions:")
print(f"F_A = {symbolic_solutions[FA]}")
print(f"F_B = {symbolic_solutions[FB]}")
```

```python
from astropy import units as u

# Define known values with units
weight_scaffold = 200 * u.N
weight_worker = 700 * u.N
length_scaffold = 3 * u.m
distance_worker = 1 * u.m

# Calculate FB using the torque equation derived in Cell 1:
# FB * L = W_w * x + W_s * (L / 2)
FB_numerical = (
    weight_worker * distance_worker + weight_scaffold * (length_scaffold / 2)
) / length_scaffold

# Calculate FA using the force equation:
# FA = W_w + W_s - FB
FA_numerical = weight_worker + weight_scaffold - FB_numerical

# Print the results with units
print("Numerical Solutions:")
print(f"Tension in the left rope (F_A): {FA_numerical.to(u.N):.1f}")
print(f"Tension in the right rope (F_B): {FB_numerical.to(u.N):.1f}")
```
