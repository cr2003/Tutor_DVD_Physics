#### Helper functions

```python
def fmt(val, decimals=2):
    """Devuelve un entero si no tiene decimales, si tiene decimales los redondea a la cantidad pedida en decimals"""
    n = float(val)
    n = round(n, decimals)
    if n == int(n):
        return f"{int(n):,}"
    else:
        return f"{n:,.{decimals}f}"
```

##  Equations of Motion in 1 Dimension

### Motion with constant acceleration

$$ x = v_0 t + \frac{1}{2}at^2 $$

$$v = v_0 + at$$

$$v^2 = v_0^2 + 2ax$$

#### Problem

A car is traveling at 60 mi/hr and the driver sees an deer 100m away on the highway.

What acceleration is needed for the car to stop before hitting the deer?

<img src="../pics/Phy1_video_2.png" width=600> 


```python
# =====================================================================
# 1. SYMBOLIC STAGE (SymPy)
# =====================================================================

# Define abstract mathematical symbols
vf, v0, a, dx = sp.symbols("v_f v_0 a Delta_x")

# Define the kinematic equation: v_f^2 = v_0^2 + 2 * a * dx
equation = sp.Eq(vf**2, v0**2 + 2 * a * dx)

# Solve the equation for the unknown acceleration 'a'
solutions = sp.solve(equation, a)
a_isolated = solutions[0]

print("--- Acceleration Formula Solved by SymPy ---")
display(Math(rf"{sp.latex(sp.Eq(a, a_isolated))}"))
```

```python
# =====================================================================
# 2. UNITS AND NUMERICAL STAGE (AstroPy)
# =====================================================================

# Assign the problem data using AstroPy units
# Note: AstroPy handles the imperial system via u.imperial
initial_velocity = 60 * u.imperial.mi / u.h
final_velocity = 0 * u.m / u.s  # The car comes to a complete stop
distance = 100 * u.m

# Convert the SymPy expression into a callable Python function
# The order of the arguments matches the symbols we want to pass
calculate_acceleration = sp.lambdify((vf, v0, dx), a_isolated)

# Evaluate the function passing the AstroPy Quantity objects directly
acceleration_result = calculate_acceleration(final_velocity, initial_velocity, distance)

# Display the final result converted to standard SI acceleration units
print("--- Final Result ---")
print(f"Required acceleration: {acceleration_result.to(u.m / u.s**2):.2f}")
```

#### Problem

A rocket is launched straight up and has an initial speed of 50 m/s. It has a constant upward acceleration of 2.00 m/s^2 until the engines stop at an altitude of 150 m.

a) What is the rocket speed at engine cutoff?

b) How high does the rocket go?

<img src="../pics/Phy1_video_2_2.png" width=500> 


```python
# a) Stage 1 until engine cutoff
#
# =====================================================================
# 1. SYMBOLIC STAGE 1 (SymPy)
# =====================================================================

# Define abstract mathematical symbols
vf, v0, a, dx = sp.symbols("v_f v_0 a Delta_x")

# Define the kinematic equation: v_f^2 = v_0^2 + 2 * a * dx
equation = sp.Eq(vf**2, v0**2 + 2 * a * dx)

# Solve the equation for the velocity at engine cutoff (vf)
solutions = sp.solve(equation, vf)
vf_isolated = solutions[1]

print("--- vf Formula Solved by SymPy ---")
display(Math(rf"{sp.latex(sp.Eq(vf, vf_isolated))}"))
```

```python
# a) Stage 1 until engine cutoff
#
# =====================================================================
# 2. UNITS AND NUMERICAL STAGE (AstroPy)
# =====================================================================

initial_velocity = 50 * (u.m / u.s)
acceleration = 2 * (u.m / u.s**2)
high_1 = 150 * u.m

# Convert the SymPy expression into a callable Python function
# The order of the arguments matches the symbols we want to pass
calculate_vf1 = sp.lambdify((v0, a, dx), vf_isolated)

# Evaluate the function passing the AstroPy Quantity objects directly
vf_result = calculate_vf1(initial_velocity, acceleration, high_1)

# Display the final result converted to standard SI acceleration units
print("--- a) Result ---")
print(f"Velocity at MECO: {vf_result:.1f}")
```

```python
# b) Stage 2 until rocket stops (max. high, vf=0)
#
# =====================================================================
# 3. SYMBOLIC STAGE 2 (SymPy)
# =====================================================================

# Define abstract mathematical symbols
vf, v0, a, dx = sp.symbols("v_f v_0 a Delta_x")

# Define the kinematic equation: v_f^2 = v_0^2 + 2 * a * dx
equation = sp.Eq(vf**2, v0**2 + 2 * a * dx)

# Solve the equation for dx when the rocket stops at max. height
solutions = sp.solve(equation, dx)[0]
dx_isolated = solutions

print("--- dx Formula Solved by SymPy ---")
display(Math(rf"{sp.latex(sp.Eq(dx, dx_isolated))}"))
```

```python
# b) Stage 2 until rocket stops (max. height, vf=0)
#
# =====================================================================
# 2. UNITS AND NUMERICAL STAGE (AstroPy)
# =====================================================================

initial_velocity = vf_result  # The velocity at cutoff from stage 1
acceleration = -const.g0  # gravity
final_velocity = 0  # Rocket stops due to gravity


# Convert the SymPy expression into a callable Python function
# The order of the arguments matches the symbols we want to pass
calculate_dx = sp.lambdify((v0, vf, a), dx_isolated)

# Evaluate the function passing the AstroPy Quantity objects directly
dx_result = calculate_dx(initial_velocity, final_velocity, acceleration)

# Display the final result converted to standard SI acceleration units
print("--- b) Result ---")
print(f"Second stage high: {dx_result:.1f}")
print(f"Total height: {high_1 + dx_result:.1f}")
```
