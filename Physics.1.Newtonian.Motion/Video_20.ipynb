#### Problem 1

A cylindrical spool with mass 0.5 kg and radius 0.5 m is wrapped with 4 m of string. A 5 kg mass is hung from the end of the string. How fast will the spool be ratating when all of the string has been pulled off?

<img src="../pics/Phy1_video_20_2.png" width=400> 


```python
# Define symbols
M, m = sp.symbols("M m")  # Mass of spool (M) and hanging mass (m)
R, h = sp.symbols("R h")  # Radius of spool (R) and length of string (h)
g, T, a = sp.symbols("g T a")  # Gravity, Tension, and linear acceleration
alpha, omega = sp.symbols(
    "alpha \omega"
)  # Angular acceleration and final angular velocity

# 1. Moment of Inertia of the spool
I = sp.Rational(1, 2) * M * R**2

# 2. Equation of motion for hanging mass: m*g - T = m*a
mass_eq = sp.Eq(m * g - T, m * a)

# 3. Equation of motion for the spool (Torque = I * alpha): T * R = I * alpha
# Since there is no slipping, a = alpha * R  => alpha = a / R
torque_eq = sp.Eq(T * R, I * (a / R))

# Solve the system of equations for linear acceleration 'a' and tension 'T'
dynamics_sol = sp.solve([mass_eq, torque_eq], (a, T))
linear_acceleration = dynamics_sol[a]

# 4. Kinematics: v^2 = 2 * a * h  => (omega * R)^2 = 2 * a * h
# Solve for omega (taking the positive root)
kinematic_eq = sp.Eq((omega * R) ** 2, 2 * linear_acceleration * h)
omega_sol = sp.solve(kinematic_eq, omega)[1]

print("Symbolic Expression for Angular Velocity (omega):")
sp.init_printing(use_unicode=True)
display(omega_sol)
```

```python
from astropy import units as u
import numpy as np

# Define known values with units
mass_spool = 0.5 * u.kg
mass_hanging = 5.0 * u.kg
radius = 0.5 * u.m
height = 4.0 * u.m
g_accel = 9.8 * u.m / u.s**2

# Calculate linear acceleration from the derived system:
# a = (m * g) / (m + 0.5 * M)
numerator_a = mass_hanging * g_accel
denominator_a = mass_hanging + 0.5 * mass_spool
a_val = numerator_a / denominator_a

# Calculate final angular velocity using kinematic relationship:
# omega = sqrt(2 * a * h) / R
omega_numerical = np.sqrt(2 * a_val * height) / radius

# Print the final result using dimensionless_angles() equivalency
print("Numerical Solution (Dynamics + Kinematics):")
print(
    f"Final angular velocity: {omega_numerical.to(u.rad / u.s, equivalencies=u.dimensionless_angles()):.2f}"
)
```
