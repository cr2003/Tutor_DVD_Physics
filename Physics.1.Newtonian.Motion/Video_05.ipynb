#### Helper functions

```python
def fmt(val, decimals=2):
    """Devuelve un entero si no tiene decimales, si tiene decimales los redondea a la cantidad pedida en decimals"""
    n = float(val)
    n = round(n, decimals)
    if n == int(n):
        return f"{int(n):,}"
    else:
        return f"{n:,.{decimals}f}"
```

##  Newtons Laws of Motion


### 1. The First Law: The Law of Inertia

An object at rest stays at rest, and an object in motion stays in motion with the same speed and in the same direction, unless acted upon by an unbalanced external force.


### 2. The Second Law: The Law of Force and Acceleration

The acceleration of an object depends directly upon the net force acting upon the object and inversely upon the mass of the object.

$$F = m a $$

Newton Units:
$$(kg)(\frac{m}{s^2}) = N$$

### 3. The Third Law: Action and Reaction

For every action, there is an equal and opposite reaction.

Forces always occur in pairs. If Object A exerts a force on Object B ($\vec{F}_{AB}$), Object B simultaneously exerts an equal and opposite force back on Object A ($\vec{F}_{BA}$).

#### Problem 1

A large mass of snow and ice breaks free from high on a mountaintop and starts a long frictionless ride to the ground. The mountain has a slope of 30 degrees and the mass of ice is initially 400 m up from the mountain base. If you are standing at the bottom, how long do you have to get out of the way?

<img src="../pics/Phy1_video_5.png" width=800> 


```python
# =====================================================================
# 1. SYMBOLIC STAGE (SymPy)
# =====================================================================

# Define abstract mathematical symbols
(
    x,
    v0,
    t,
    a,
) = sp.symbols("x, v0, t, a", positive=True)

# Define the kinematic equation: x = v_0 * t + 1/2 a * t^2
# Since the initial velocity is zero, we have
equation = sp.Eq(x, sp.Rational(1, 2) * a * t**2)

# Solve the equation for the unknown time 't'
solutions = sp.solve(equation, t)
t_isolated = solutions[0]

print("--- time Formula Solved by SymPy ---")
display(Math(rf"{sp.latex(sp.Eq(t, t_isolated))}"))
```

```python
# =====================================================================
# 2. UNITS AND NUMERICAL STAGE (Pure AstroPy + NumPy)
# =====================================================================

# 1. Define your inputs with AstroPy units
distance_x = 400 * u.m
# From the figure a = 1/2 g
acceleration = 1 / 2 * const.g0

t_result = np.sqrt(2 * distance_x / acceleration)

print("--- a) t result ---")
print(f"Time to get out: {t_result:.1f}")
```

#### Problem 2

A 5 g bullet leaves the muzzle of a riffle with a speed of 320 m/s. The length of the barrel is 0.82 m. What force is exerted on the bullet while it is traveling down the barrel?  

<img src="../pics/Phy1_video_5_2.png" width=800> 


```python
# =====================================================================
# 1. SYMBOLIC STAGE (SymPy)
# =====================================================================

# Define abstract mathematical symbols
(
    x,
    v,
    v0,
    a,
) = sp.symbols("x, v, v0, a", positive=True)

# Define the kinematic equation: v^2 = v_0^2 + 2 a x
# Since the initial velocity is zero, we have
equation = sp.Eq(v**2, 2 * a * x)

# Solve the equation for the unknown 'a'
solutions = sp.solve(equation, a)
a_isolated = solutions[0]

print("--- time Formula Solved by SymPy ---")
display(Math(rf"{sp.latex(sp.Eq(a, a_isolated))}"))
```

```python
# =====================================================================
# 2. UNITS AND NUMERICAL STAGE (Pure AstroPy + NumPy)
# =====================================================================

# 1. Define your inputs with AstroPy units
distance_x = 0.82 * u.m
mass = 5 * u.g
vel = 320 * (u.m / u.s)

accel = vel**2 / (2 * distance_x)
F = mass * accel

print("--- a) F result ---")
print(f"Force: {F.to(u.N):.0f}")
```

#### Problem 3

Find the tension in the two wires that support the light fixture. The light wighs 100 N.  

<img src="../pics/Phy1_video_5_3.png" width=800> 



```python
# =====================================================================
# 1. SYMBOLIC STAGE (SymPy)
# =====================================================================

# Define abstract mathematical symbols
(
    T_1,
    theta,
    w,
) = sp.symbols("T_1, theta, w", positive=True)

# Since the tension in the wires is the same, T_1 = T_2
# Define the Equilibrium equation: T_1 sin(theta) + T_1 sin(theta) - 100 = 0

equation = sp.Eq(0, T_1 * sp.sin(theta) + T_1 * sp.sin(theta) - w)

# Solve the equation for the unknown T_1
solutions = sp.solve(equation, T_1)[0]
T_1_isolated = solutions

print("--- time Formula Solved by SymPy ---")
display(Math(rf"{sp.latex(sp.Eq(T_1, T_1_isolated))}"))
```

```python
# =====================================================================
# 2. UNITS AND NUMERICAL STAGE (Pure AstroPy + NumPy)
# =====================================================================

# 1. Define your inputs with AstroPy units
weight = 100 * u.N
theta_val = 40 * u.deg

Tension_1 = weight / (2 * np.sin(theta_val))

print("--- a) Wire Tension result ---")
print(f"Wire tension: {Tension_1:.0f}")
```

#### Problem 4

A 2000 kg sailboat has an 3000 N eastward force exerted on it by the ocean waves. The wind also blows against the sails with a force of 6000 N, directed 45 degrees north of west. What is the magnitude of the resultant acceleration?

<img src="../pics/Phy1_video_5_4.png" width=800> 



```python
# =====================================================================
# UNITS AND NUMERICAL STAGE (Pure AstroPy + NumPy)
# =====================================================================

# 1. Define your inputs with AstroPy units
mass = 2000 * u.kg
F1 = 3000 * u.N
F2 = 6000 * u.N
theta_val = 45 * u.deg

Fx = -F2 * np.cos(theta_val) + F1
Fy = F2 * np.sin(theta_val)
F_mag = np.sqrt(Fx**2 + Fy**2)

# F = m a ==> a = F / m
accel = F_mag / mass

print(f"Magnitude resultant acceleration: {accel.to(u.m / u.s**2):.1f}")
```
