### Pressure $P$
Units: Pascal

#### Problem 1

A pìpe is filled water with a pressure of $5\times 10^5$ above atmospheric pressure. A 4 mm diameter hole developes in the pipe. How much force must you exert on the hole in order to stop the leak?

<img src="../pics/Phy1_video_22.png" width=400> 


```python
# Define known values with units
r = (4 / 2 * u.mm).to(u.m)
P = 5e5 * u.Pa

# calculate the area of the hole: A = pi * r^2
A = np.pi * r**2

# calculate the force using pressure formula: P = F / A ==> F = P * A
F = P * A
print(f"Force: {F.to(u.N):.2f}")
```
