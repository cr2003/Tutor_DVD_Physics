##  Kinetic Energy

$$KE = \frac{1}{2}m v^2 $$

Units of Jules [J]

### Work - Energy Theorem

$$W_{net} = KE_f - KE_i = \Delta KE$$

$$W_{net} = \frac{1}{2}m v_f^2 - \frac{1}{2}m v_i^2$$





#### Problem 1

A 0.150 kg baseball is thrown at a speed of 40 m/s at an initial angle of 30 deg. What is kinetic energy of the ball at he highest point in its motion?

<img src="../pics/Phy1_video_8.png" width=600> 


```python
# -----------------------------------------------------------------
# Numerical Evaluation with AstroPy Units
# -----------------------------------------------------------------

m = 0.15 * u.kg
v0 = 40 * (u.m / u.s)
theta = 30 * u.deg

# Calculate constant v_x
v_x = v0 * np.cos(theta)

# At apex, v_y = 0, so v = v_x
v = v_x
KE = 1 / 2 * m * v**2

print(f"KE: {KE.to(u.J):.0f} ")
```

#### Problem 2

A particle has a mass of 0.60 kg and a speed of 2 m/s at point A. At point B, the same particle has a kinetic energy of 7.5 J.

a) What is the KE at A?

b) Speed at point B?

c) Total work done on the particle as it moves from A to B?

<img src="../pics/Phy1_video_8_2.png" width=600> 



```python
# -----------------------------------------------------------------
# Numerical Evaluation with AstroPy Units
# -----------------------------------------------------------------

m = 0.60 * u.kg
v_0 = 2 * (u.m / u.s)
KE_B = 7.5 * u.J

# a)
KE_A = 1 / 2 * m * v_0**2

print(f"KE at point A: {KE_A.to(u.J):,.1f}")

# b)
# KE_B = 1/2 * m * v_B**2 ==> v_B = sqrt(2*KE_B/m)
v_B = np.sqrt(2 * KE_B / m)
print(f"Speed at point B: {v_B.to(u.m / u.s):.0f}")
# c)
delta_w = KE_B - KE_A
print(f"Total work: {delta_w:.1f}")
```

#### Problem 3

A bowling ball has a mass of 7 kg and moves at a speed 3 m/s. If a ping-pong ball has a mass of 2.45 g, how fast must it move in order to have the same KE as the bowling ball?


<!-- <img src="../pics/Phy1_video_8_2.png" width=600>  -->



```python
# -----------------------------------------------------------------
# Numerical Evaluation with AstroPy Units
# -----------------------------------------------------------------

Bowling_mass = 7 * u.kg
Bowling_v = 3 * (u.m / u.s)
ping_pong_mass = 2.45 * u.g

KE_Bowling = 1 / 2 * Bowling_mass * Bowling_v**2
KE_ping_pong = KE_Bowling

# KE_ping_pong = 1/2 * ping_pong_mass * ping_pong_v**2 ==> ping_pong_v = sqrt(2 * KE_ping_pong / ping_pong_mass )

ping_pong_v = np.sqrt(2 * KE_ping_pong / ping_pong_mass)
print(f"Ping-pong velocity: {ping_pong_v.to(u.m / u.s):.0f}")
```
