#### Problem 1

A person swings a golf club at a golf ball. The head of the club is 200 g and it is traveling at 55 m/s just prior to striking the golf ball. The ball is 46 g and is at rest prior to the collision. After the collision the golf club is traveling at 40 m/s. What is the speed of the ball just after the collision.





```python
# -----------------------------------------------------------------
# 2. Numerical Evaluation with AstroPy Units
# -----------------------------------------------------------------

# Given parameters with AstroPy units
mass_club = 200 * u.g
vel_club_i = 55 * (u.m / u.s)
mass_ball = 46 * u.g
vel_ball_i = 0 * (u.m / u.s)
vel_club_f = 40 * (u.m / u.s)

# momentum:  mass_club * vel_club_i + mass_ball * vel_ball_i = mass_club * vel_club_f + mass_ball * vel_ball_f
vel_ball_f = (
    (mass_club * vel_club_i) + (mass_ball * vel_ball_i) - (mass_club * vel_club_f)
) / mass_ball
print("--- AstroPy Numerical Result ---")
print(f"Ball velocity: {vel_ball_f.to(u.m / u.s):.1f}")
```

#### Problem 2

An astronaut (mass 60 kg) is on a pacewalk and her tether breaks. Se can throw her 10 kg oxygen tank away from her at a velocity of 12 m/s in an effort to porpel herself back to the ship. How far can she be from the ship when the line breaks such that she can return  within 60 seconds?

<img src="../pics/Phy1_video_12_2.png" width=800> 


```python
# -----------------------------------------------------------------
# 2. Numerical Evaluation with AstroPy Units
# -----------------------------------------------------------------

# Given parameters with AstroPy units
mass_astronaut = 60 * u.kg
mass_tank = 10 * u.kg
vel_tank_i = 0 * (u.m / u.s)
vel_tank_f = -12 * (u.m / u.s)  # Tank throwing away to the left, x-axis
vel_astronaut_i = 0 * (u.m / u.s)
time_limit = 60 * u.s


# momentum:  mass_astronaut * vel_astronaut_i + mass_tank * vel_tank_i = mass_astronau * vel_astronaut_f + mass_tank * vel_tank_f
vel_astronaut_f = (
    (mass_astronaut * vel_astronaut_i)
    + (mass_tank * vel_tank_i)
    - (mass_tank * vel_tank_f)
) / mass_astronaut

# Distance Limit
# v = d / t ==> d = v * t
distance_limit = vel_astronaut_f * time_limit
print("--- AstroPy Numerical Result ---")
print(f"Distance limit: {distance_limit:.0f}")
```
